```python
import mysql.connector
# import numpy as np
import pandas as pd
```

```python
db_connection = mysql.connector.connect(user="greg", password="greg")
db_cursor = db_connection.cursor()
db_cursor.execute("CREATE DATABASE TestDB;")
db_cursor.execute("USE TestDB;")
```

```python
db_cursor.execute("CREATE TABLE Wines(fixed_acidity FLOAT, volatile_acidity FLOAT, citric_acid FLOAT, \
                   residual_sugar FLOAT, chlorides FLOAT, free_so2 FLOAT, total_so2 FLOAT, density FLOAT, \
                   pH FLOAT, sulphates FLOAT, alcohol FLOAT, quality INT, is_red INT);")
```

```python
red_wines = pd.read_csv("winequality-red.csv", sep=";")
red_wines["is_red"] = 1
white_wines = pd.read_csv("winequality-white.csv", sep=";")
white_wines["is_red"] = 0
all_wines = pd.concat([red_wines, white_wines])
```

```python
all_wines
```

```python
wine_tuples = list(all_wines.itertuples(index=False, name=None))
wine_tuples_string = ",".join(["(" + ",".join([str(w) for w in wt]) + ")" for wt in wine_tuples])
wine_tuples_string[:100]
```

```python
db_cursor.execute("INSERT INTO Wines(fixed_acidity, volatile_acidity, citric_acid, residual_sugar, \
                   chlorides, free_so2, total_so2, density, pH, sulphates, alcohol, quality, is_red) \
                   VALUES " + wine_tuples_string + ";")
```

```python
db_cursor.execute("FLUSH TABLES;")
```
